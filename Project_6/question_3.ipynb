```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras import layers, models, regularizers
from tensorflow.keras.datasets import boston_housing
from sklearn.model_selection import KFold
import matplotlib.pyplot as plt


(x_train, y_train), (x_test, y_test) = boston_housing.load_data()
mean = np.mean(x_train, axis=0)
std = np.std(x_train, axis=0)

x_train = (x_train - mean) / std
x_test = (x_test - mean) / std
```

```python
# Function to build the model
def build_model(hidden_units, regularization=False):
    model = models.Sequential()
    model.add(layers.InputLayer(input_shape=(x_train.shape[1],)))
    
    if isinstance(hidden_units, tuple):  # same as earlier where I consider 2 hidden layers
        for units in hidden_units:
            model.add(layers.Dense(units, activation='relu', 
                                   kernel_regularizer=regularizers.l2(0.001) if regularization else None))
    else:  # If 1 hidden layer
        model.add(layers.Dense(hidden_units, activation='relu', 
                               kernel_regularizer=regularizers.l2(0.001) if regularization else None))
    
    model.add(layers.Dense(1))  # Output layer with no activation
    model.compile(optimizer='adam', loss='mae')  # MAE for regression
    
    return model
```

```python
# first we build the model with 2 hidden layers, each with 64 units, and 200 epochs
model_a = build_model((64, 64)) 
history_a = model_a.fit(x_train, y_train, epochs=200, batch_size=16, validation_split=0.2, verbose=0)

# Plot validation MAE against epochs for model to assess the optimal epochs
plt.plot(history_a.history['val_loss'], label='Validation MAE')
plt.title('Validation MAE vs Epochs for Model (a)')
plt.xlabel('Epochs')
plt.ylabel('Validation MAE')
plt.legend()
plt.show()

optimal_epochs = 25 


model_a = build_model((64, 64))  # Rebuild the model to avoid overfitting
#each model below can be built in a similar manner
history_a_optimal = model_a.fit(x_train, y_train, epochs=optimal_epochs, batch_size=16, validation_split=0.2, verbose=0)
val_mae_a = min(history_a_optimal.history['val_loss'])  # Best validation MAE
print(f"Model (a) - Validation MAE: {val_mae_a:.4f}")

# (b) Fiting a neural network model with 1 hidden layer with 128 units
model_b = build_model(128)
history_b = model_b.fit(x_train, y_train, epochs=optimal_epochs, batch_size=16, validation_split=0.2, verbose=0)
val_mae_b = min(history_b.history['val_loss'])
print(f"Model (b) - Validation MAE: {val_mae_b:.4f}")

# (c) Adding L2 weight regularization to the model 
model_c = build_model((64, 64), regularization=True)
history_c = model_c.fit(x_train, y_train, epochs=optimal_epochs, batch_size=16, validation_split=0.2, verbose=0)
val_mae_c = min(history_c.history['val_loss'])
print(f"Model (c) - Validation MAE with L2 regularization: {val_mae_c:.4f}")

# (d) Adding L2 weight regularization/1 hidden layer with 128 hidden units
model_d = build_model(128, regularization=True)
history_d = model_d.fit(x_train, y_train, epochs=optimal_epochs, batch_size=16, validation_split=0.2, verbose=0)
val_mae_d = min(history_d.history['val_loss'])
print(f"Model (d) - Validation MAE with L2 regularization: {val_mae_d:.4f}")
```

```python
# last testing for model c mae (indiviudal adjustment)
test_mae_c = model_c.evaluate(x_test, y_test, verbose=0)
print(f"Recommended Model (c) - Test MAE: {test_mae_c:.4f}")
```
